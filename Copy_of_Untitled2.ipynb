Visualization a Few Features

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['tweet_id', 'author_id', 'inbound', 'created_at', 'text']  # Categories on X-axis
values = [5, 7, 3, 8, 6]               # Values on Y-axis

# Create bar chart
plt.bar(categories, values, color='skyblue')

# Add title and labels
plt.title('Sample Bar Graph')
plt.xlabel('Categories')
plt.ylabel('Values')

# Display chart
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a grid of x, y points
Y, X = np.mgrid[-3:3:100j, -3:3:100j]

# Define vector field (U = x-component, V = y-component)
U = -1 - X**2 + Y
V = 1 + X - Y**2

# Create the streamplot
plt.streamplot(X, Y, U, V, color=U, linewidth=1, cmap='autumn')

# Add title and labels
plt.title('Flow Line (Stream) Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Display plot
plt.show()
```

Model building

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Sample data
X = np.array([[1], [2], [3], [4], [5]])  # Features
y = np.array([2, 4, 5, 4, 5])            # Target

# Create and train model
model = LinearRegression()
model.fit(X, y)

# Predict
y_pred = model.predict(X)

# Print model parameters
print("Coefficient:", model.coef_)
print("Intercept:", model.intercept_)

# Plot original data and prediction
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, y_pred, color='red', label='Predicted')
plt.title('Linear Regression Model')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

Evaluation

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Actual and predicted values
y_true = np.array([3, -0.5, 2, 7])
y_pred = np.array([2.5, 0.0, 2, 8])

# Evaluation metrics
mse = mean_squared_error(y_true, y_pred)
r2 = r2_score(y_true, y_pred)

print("Mean Squared Error:", mse)
print("R-squared Score:", r2)
```

Predict the final code

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Step 1: Sample dataset
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 5, 4, 5])

# Step 2: Model building and training
model = LinearRegression()
model.fit(X, y)

# Step 3: Prediction
y_pred = model.predict(X)

# Step 4: Evaluation
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
print("Model Evaluation:")
print("Mean Squared Error (MSE):", mse)
print("R-squared (R2) Score:", r2)
```
